```python
# displaying the relationship between the hours_per_day spent on music and the levels of anxiety and depression in individuals.
plt.figure(figsize=(10, 6))
plt.scatter(music_metadata['hours_per_day'], music_metadata['anxiety'], color='blue', label='Anxiety')
plt.scatter(music_metadata['hours_per_day'], music_metadata['depression'], color='red', label='Depression')
```

The scatter plot of hours_per_day vs. anxiety (blue dots) represents the relationship between the time spent on music and anxiety levels.
The scatter plot of hours_per_day vs. depression (red dots) represents the relationship between the time spent on music and depression levels.
By plotting these two relationships on the same graph, we visualize and compare the correlations between music engagement and mental health outcomes (anxiety and depression).


```python


import seaborn as sns
import matplotlib.pyplot as plt

# Create a new column in the dataframe to specify the hue
music_metadata['condition'] = ['Anxiety'] * len(music_metadata)

# Use Seaborn's lmplot function to create the plot
sns.lmplot(x='hours_per_day', y='anxiety', data=music_metadata, hue='condition')

# Create a new column in the dataframe to specify the hue
music_metadata['condition'] = ['Depression'] * len(music_metadata)

# Use Seaborn's lmplot function to create the plot
sns.lmplot(x='hours_per_day', y='depression', data=music_metadata, hue='condition')

# Show the plot
plt.show()
```
